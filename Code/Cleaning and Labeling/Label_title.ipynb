```python
import pandas as pd
import numpy as np
```

```python
# Load data
Title_Train = pd.read_csv('Title_Train.csv')
Title_Train
```

```python
# Load green and non green titles
green_titles = pd.read_csv('GreenTitles.csv', usecols=['Titles'])
green_titles

non_green_titles = pd.read_csv('NonGreenTitles.csv', usecols=['Titles'])
non_green_titles
```

```python
# Drop title duplicates if any (usually not)
Title_Train = Title_Train.drop_duplicates(subset=['title'], ignore_index = True)
```

```python
# Label data
green = np.zeros(Title_Train['title_tokens'].shape[0])

for i in range(Title_Train['title_tokens'].shape[0]):
    Title_Train['title_tokens'].iloc[i]=Title_Train['title_tokens'].iloc[i].replace('[', '').replace(']', '').replace("'", '')
    for j in range(green_titles['Titles'].shape[0]):
        if all(word in Title_Train['title_tokens'].iloc[i].split(', ') for word in green_titles['Titles'].iloc[j].split(' ')):
            green[i]=1
        
    for j in range(non_green_titles.shape[0]):
        if all(word in Title_Train['title_tokens'].iloc[i].split(', ') for word in non_green_titles['Titles'].iloc[j].split(' ')) & (green[i]!=1):
            green[i]=-1
            
        elif all(word in Title_Train['title_tokens'].iloc[i].split(', ') for word in non_green_titles['Titles'].iloc[j].split(' ')) & (green[i]==1):
            print("Ambiguous title set to 0 :")
            print(Title_Train['title'].iloc[i])
            green[i]=0
            

Title_Train['green_title'] = green
```

```python
# Save labeled data 
Title_Train[['title', 'green_title']].to_csv('Title_Train_Label.csv')
```
